```python
%matplotlib inline

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set(style="ticks", context="notebook", font_scale=1.2) # 

df = sns.load_dataset("anscombe") #  anscombe  dataframe 

# lmplot  lmplot 
sns.lmplot(
    data=df,  # DataFrame 
    x='x',  # x 
    y='y',  # y 
    col='dataset',  # 
    hue='dataset',  # 
    col_wrap=2,  #  col 
    ci=None,  #  None  None
    palette='muted',  # 
    height=4,  # 
    aspect=1.5,  # 
    truncate=True,  #  x 
    markers=["o", "s", "D", "^"],  # 
    scatter_kws={"s":50, "alpha":1},  #  plt.scatter 
    line_kws={"color":"red", "linewidth":2}  #  plt.plot 
)
```

```python
"""  """
sns.set()
iris = sns.load_dataset("iris")

g = sns.lmplot(data=iris, x="sepal_length", y="sepal_width", hue="species",
               truncate=True, height=5)
g.set_axis_labels("Sepal length(mm)", "Sepal width(mm)")
```

```python
"""  """
sns.set(style='darkgrid')
titanic = sns.load_dataset("titanic")

pal = dict(male="#6495ED", female="#F08080")  # 

g = sns.lmplot(  # 
    data=titanic,
    x='age',
    y='survived',
    col='sex',
    hue='sex',
    palette=pal,
    y_jitter=.02,
    # logistic=True
)
g.set(xlim=(0, 80), ylim=(-.05, 1.05))  #  x  [0, 80], y  [-0.05, 1.05]
plt.show()
```
