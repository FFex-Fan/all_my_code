```python
%matplotlib inline

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set(style="ticks", context="notebook", font_scale=1.2) # 全局设置绘图样式

df = sns.load_dataset("anscombe") # 加载 anscombe 数据集，为 dataframe 类型

# lmplot 绘制回归图，通过 lmplot 可以直观的总揽数据内部关系
sns.lmplot(
    data=df,  # 输入的数据集（DataFrame 格式）。
    x='x',  # 自变量（x 轴）。（字符串）
    y='y',  # 因变量（y 轴）。（字符串）
    col='dataset',  # 变量用于列子图分组。（字符串）
    hue='dataset',  # 变量用于分组和颜色编码。（字符串）
    col_wrap=2,  # 设置每行显示的列数（当使用 col 分组时）。（整数）
    ci=None,  # 置信区间大小（以百分比表示），如果为 None 则不绘制置信区间。（整数或 None）
    palette='muted',  # 颜色调色板。（字符串或列表）
    height=4,  # 每个子图的高度（英寸）。（浮点数）
    aspect=1.5,  # 设置每个子图的宽高比。（浮点数）
    truncate=True,  # 回归线在数据的 x 范围内截断。
    markers=["o", "s", "D", "^"],  # 数据点的标记样式。（列表或字符串）
    scatter_kws={"s":50, "alpha":1},  # 传递给 plt.scatter 的关键字参数，用于自定义散点图。（字典）
    line_kws={"color":"red", "linewidth":2}  # 传递给 plt.plot 的关键字参数，用于自定义回归线（字典）
)
```

```python
""" 多元线性回归 """
sns.set()
iris = sns.load_dataset("iris")

g = sns.lmplot(data=iris, x="sepal_length", y="sepal_width", hue="species",
               truncate=True, height=5)
g.set_axis_labels("Sepal length(mm)", "Sepal width(mm)")
```

```python
""" 有小面的逻辑回归 """
sns.set(style='darkgrid')
titanic = sns.load_dataset("titanic")

pal = dict(male="#6495ED", female="#F08080")  # 制作一个带有性别色彩的自定义调色板

g = sns.lmplot(  # 显示生存能力与年龄和性别的关系
    data=titanic,
    x='age',
    y='survived',
    col='sex',
    hue='sex',
    palette=pal,
    y_jitter=.02,
    # logistic=True
)
g.set(xlim=(0, 80), ylim=(-.05, 1.05))  # 设置 x 轴的范围为 [0, 80], y 轴的范围为 [-0.05, 1.05]
plt.show()
```
